```python
#Assumption you have a 10 elements List
# indexes here are 0 to 9
from typing import List
def tr_ind(ind, i) -> int:
    # Assumption here we dropped virtuall item at index ind  
    # The virtual new list now runs from 0 to a max index of 8
    # Given the position i in the virtual list.. can you give me the position of the original list
    # i starts from 0 to an 8 in out case
    if i >= ind: return i+1
    else: return i
l: List[int] = [-4,-1,-1,0,1,2]   
print (f"Drop item at position 0 ; new list of size 5.. Item at position 0 of new list = {l[tr_ind(0,0)]}")
print (f"Drop item at position 0 ; new list of size 5.. Item at position 4 of new list = {l[tr_ind(0,4)]}")
print (f"Drop item at position 1 ; new list of size 5.. Item at position 0 of new list = {l[tr_ind(1,0)]}")
print (f"Drop item at position 1 ; new list of size 5.. Item at position 1 of new list = {l[tr_ind(1,1)]}")
print (f"Drop item at position 1 ; new list of size 5.. Item at position 2 of new list = {l[tr_ind(1,2)]}")
print (f"Drop item at position 1 ; new list of size 5.. Item at position 4 of new list = {l[tr_ind(1,4)]}")
print (f"Drop item at position 5 ; new list of size 5.. Item at position 0 of new list = {l[tr_ind(5,0)]}")
print (f"Drop item at position 5 ; new list of size 5.. Item at position 4 of new list = {l[tr_ind(5,4)]}")
```
