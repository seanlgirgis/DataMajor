### Problem Description
> Given a string `s` containing just the characters `'('`, `')'`, `'{'`, `'}'`, `'['` and `']'`, determine if the input string is valid.
> 

```python
class Solution(object):
    def isValid(self, s: str) -> bool:
        pairs = {')': '(', '}': '{', ']': '['}   # â† reversed mapping!
        stack = []
        
        for ch in s:
            if ch in '({[':
                stack.append(ch)
            elif ch in ')}]':
                if not stack or stack.pop() != pairs[ch]:
                    return False
            else:
                return False  
        return not stack
sol = Solution()
print("Test1: s='()' -> True: success" if sol.isValid("()") == True else "Test1: Fail")
print("Test2: s='()[]{}' -> True: success" if sol.isValid("()[]{}") == True else "Test2: Fail")
print("Test3: s='(]' -> False: success" if sol.isValid("(]") == False else "Test3: Fail")
print("Test4: s=']' (edge case single close) -> False: success" if sol.isValid("]") == False else "Test4: Fail")
```

```python
# The Test Suite: (input_string, expected_bool)
test_cases = [
    ("()", True),               # Basic pair
    ("()[]{}", True),           # Multiple pairs
    ("(]", False),              # Mismatched pair
    ("([)]", False),            # Incorrect nesting
    ("{[]}", True),             # Correct nesting
    ("]", False),               # Orphaned closer (Test 4)
    ("(", False),               # Unclosed opener
    ("", True),                 # Empty string (technically valid)
    ("((()))", True),           # Deep nesting
    ("(){", False),             # Extra opener at end
    ("abc", False)              # Invalid characters (covered by your 'else')
]

print(f"{'Input String':<15} | {'Exp':<6} | {'Act':<6} | {'Status'}")
print("-" * 45)

sol = Solution()

for s, expected in test_cases:
    try:
        actual = sol.isValid(s)
        status = "âœ… PASS" if actual == expected else "âŒ FAIL"
        print(f"{s:<15} | {str(expected):<6} | {str(actual):<6} | {status}")
    except Exception as e:
        print(f"{s:<15} | {str(expected):<6} | {'ERR':<6} | ðŸ’¥ {type(e).__name__}")
```
