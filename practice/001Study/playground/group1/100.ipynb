> Given the roots of two binary trees p and q, write a function to check if they are the same or not. Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

```python
from typing import Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if p is None and q is None : return True           # we have reached the end of the trees
        if p is None or q is None : return False           # One is exhausted and the other not yet Not matching
        if p.val != q.val :return False                    # Values are not mathcing. Trees are not the same
        return (self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right))



sol = Solution()
p1 = TreeNode(1, TreeNode(2), TreeNode(3))
q1 = TreeNode(1, TreeNode(2), TreeNode(3))
print("Test1: [1,2,3], [1,2,3] -> True: success" if sol.isSameTree(p1, q1) == True else "Test1: Fail")

p2 = TreeNode(1, TreeNode(2))
q2 = TreeNode(1, None, TreeNode(2))
print("Test2: [1,2], [1,null,2] (edge case) -> False: success" if sol.isSameTree(p2, q2) == False else "Test2: Fail")

p3 = TreeNode(1, TreeNode(2), TreeNode(1))
q3 = TreeNode(1, TreeNode(1), TreeNode(2))
print("Test3: [1,2,1], [1,1,2] -> False: success" if sol.isSameTree(p3, q3) == False else "Test3: Fail")
```

```python
from typing import Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Helper to build trees from list: [1, 2, 3] -> Tree
def build_tree(nodes: list) -> Optional[TreeNode]:
    if not nodes: return None
    root = TreeNode(nodes[0])
    queue = [root]
    i = 1
    while queue and i < len(nodes):
        node = queue.pop(0)
        if i < len(nodes) and nodes[i] is not None:
            node.left = TreeNode(nodes[i])
            queue.append(node.left)
        i += 1
        if i < len(nodes) and nodes[i] is not None:
            node.right = TreeNode(nodes[i])
            queue.append(node.right)
        i += 1
    return root

# The Test Suite: ((list_p, list_q), expected_bool)
test_cases = [
    (([1, 2, 3], [1, 2, 3]), True),           # Identical
    (([1, 2], [1, None, 2]), False),         # Different structure
    (([1, 2, 1], [1, 1, 2]), False),         # Same values, different order
    (([], []), True),                        # Both empty
    (([1], []), False),                      # One empty
    (([1, 2], [1, 3]), False),               # Different values
    (([10, 5, 15], [10, 5, 15]), True),      # Standard BST shape
    (([1, None, 2, None, 3], [1, None, 2, None, 3]), True) # Skewed trees
]

print(f"{'Tree P':<15} | {'Tree Q':<15} | {'Exp':<6} | {'Act':<6} | {'Status'}")
print("-" * 65)

sol = Solution()

for (list_p, list_q), expected in test_cases:
    p = build_tree(list_p)
    q = build_tree(list_q)
    try:
        actual = sol.isSameTree(p, q)
        status = "âœ… PASS" if actual == expected else "âŒ FAIL"
        print(f"{str(list_p):<15} | {str(list_q):<15} | {str(expected):<6} | {str(actual):<6} | {status}")
    except Exception as e:
        print(f"{str(list_p):<15} | {str(list_q):<15} | {str(expected):<6} | {'ERR':<6} | ðŸ’¥ {type(e).__name__}")
```
