### Problem Description
> Given two strings `s` and `t`, return `true` if `t` is an anagram of `s`, and `false` otherwise. An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

```python
class Solution(object):
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t): return False
        thash = {}
        shash = {}
        for ch in s: shash[ch] = shash.get(ch,0) + 1
        for ch in t: thash[ch] = thash.get(ch,0) + 1
        return shash == thash
            


sol = Solution()
print("Test1: s='anagram', t='nagaram' -> True: success" if sol.isAnagram("anagram", "nagaram") == True else "Test1: Fail")
print("Test2: s='rat', t='car' -> False: success" if sol.isAnagram("rat", "car") == False else "Test2: Fail")
print("Test3: s='a', t='ab' (edge case length mismatch) -> False: success" if sol.isAnagram("a", "ab") == False else "Test3: Fail")
```

```python
# The Test Suite: ((string_s, string_t), expected_bool)
test_cases = [
    (("anagram", "nagaram"), True),    # Classic anagram
    (("rat", "car"), False),           # Same length, different characters
    (("a", "ab"), False),              # Length mismatch
    (("", ""), True),                  # Empty strings
    (("aacc", "ccac"), False),         # Same chars, different frequencies
    (("awesome", "awesome"), True),    # Identical strings
    (("ab", "a"), False),              # Length mismatch (t shorter)
    (("abc", "def"), False),           # Completely different
    (("aabbcc", "abcabc"), True)       # Complex shuffle
]

print(f"{'String S':<12} | {'String T':<12} | {'Exp':<6} | {'Act':<6} | {'Status'}")
print("-" * 55)

sol = Solution()

for (s, t), expected in test_cases:
    try:
        actual = sol.isAnagram(s, t)
        status = "âœ… PASS" if actual == expected else "âŒ FAIL"
        print(f"{s:<12} | {t:<12} | {str(expected):<6} | {str(actual):<6} | {status}")
    except Exception as e:
        print(f"{s:<12} | {t:<12} | {str(expected):<6} | {'ERR':<6} | ðŸ’¥ {type(e).__name__}")
```
