> Given the `root` of a binary tree, return its maximum depth. A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.


```python
from typing import Optional
r"""
Thinking: -
    - Recursive wins
    - if object is None .. return 0   (end of a tree)
    - return max (maxDepth (left and right ) + 1
    

"""
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        #No run forever
        if root is None:
            return 0
        return max(self.maxDepth(root.left) , self.maxDepth(root.right)) + 1


sol = Solution()
root1 = TreeNode(3, TreeNode(9), TreeNode(20, TreeNode(15), TreeNode(7)))
print("Test1: [3,9,20,null,null,15,7] -> 3: success" if sol.maxDepth(root1) == 3 else "Test1: Fail")

root2 = TreeNode(1, None, TreeNode(2))
print("Test2: [1,null,2] -> 2: success" if sol.maxDepth(root2) == 2 else "Test2: Fail")

root3 = None
print("Test3: [] (edge case) -> 0: success" if sol.maxDepth(root3) == 0 else "Test3: Fail")
```

```python
from typing import Optional

# Helper to build trees from list: [3, 9, 20, None, None, 15, 7] -> Tree
def build_tree(nodes: list) -> Optional[TreeNode]:
    if not nodes: return None
    root = TreeNode(nodes[0])
    queue = [root]
    i = 1
    while queue and i < len(nodes):
        node = queue.pop(0)
        if i < len(nodes) and nodes[i] is not None:
            node.left = TreeNode(nodes[i])
            queue.append(node.left)
        i += 1
        if i < len(nodes) and nodes[i] is not None:
            node.right = TreeNode(nodes[i])
            queue.append(node.right)
        i += 1
    return root

# The Test Suite: (list_representation, expected_depth)
test_cases = [
    ([3, 9, 20, None, None, 15, 7], 3),  # Standard balanced tree
    ([1, None, 2], 2),                   # Right-skewed
    ([1, 2, None], 2),                   # Left-skewed
    ([], 0),                             # Empty tree (Edge case)
    ([0], 1),                            # Single node
    ([1, 2, 3, 4, 5], 3),                # Full tree
    ([1, None, 2, None, 3, None, 4], 4)  # Linked list style (Deep)
]

print(f"{'Tree List':<30} | {'Exp':<5} | {'Act':<5} | {'Status'}")
print("-" * 55)

sol = Solution()

for nodes, expected in test_cases:
    root = build_tree(nodes)
    try:
        actual = sol.maxDepth(root)
        status = "âœ… PASS" if actual == expected else "âŒ FAIL"
        print(f"{str(nodes):<30} | {expected:<5} | {actual:<5} | {status}")
    except Exception as e:
        print(f"{str(nodes):<30} | {expected:<5} | {'ERR':<5} | ðŸ’¥ {type(e).__name__}")
```
