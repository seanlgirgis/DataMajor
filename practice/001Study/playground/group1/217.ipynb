> Given an integer array `nums`, return `true` if any value appears at least twice in the array, and return `false` if every element is distinct.


```python
# Examples
# list, result
[1, 2, 3, 1], True
[1, 2, 3, 4], False
[1, 1, 1, 3, 3, 4, 3, 2, 4, 2], True
[], False
[0], False
[-1, -1], True
```

```python
from typing import List
def contains_duplicate(nums:List[int]) -> bool:
    seen = set()
    for num in nums:
        if num in seen:
            return True
        else:
            seen.add(num)
    return False
print ("Single Test Success" if contains_duplicate([1, 2, 3, 4]) == False else "Single Test Fail")

    
```

```python
# Test Cases
test_data = [
    ([1, 2, 3, 1], True),
    ([1, 2, 3, 4], False),
    ([1, 1, 1, 3, 3, 4, 3, 2, 4, 2], True),
    ([], False),
    ([0], False),
    ([-1, -1], True)
]

# Execution Loop
print(f"{'Input':<40} | {'Expected':<10} | {'Result':<10} | {'Status'}")
print("-" * 80)

for nums, expected in test_data:
    actual = contains_duplicate(nums)
    status = "✅ PASS" if actual == expected else "❌ FAIL"
    print(f"{str(nums):<40} | {str(expected):<10} | {str(actual):<10} | {status}")
```
