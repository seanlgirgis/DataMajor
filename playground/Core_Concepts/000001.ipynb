```python
print(name := "alice")
print(name.upper()    )      # "ALICE"
print(name.capitalize()  )   # "Alice"
print(name.replace("a", "o")) # "olice"

print(sentence := "  too much space  ")
print(sentence.strip()  )    # "too much space"

print(csv := "a,b,c,d")
print(csv.split(",")    )    # ["a", "b", "c", "d"]
```

```python
print(word := "Python")
print(word[0])      # "P"
print(word[-1])     # "n"  (negative = from the end)
print(word[0:3] )   # "Pyt" (up to but not including index 3)
print(word[::2] )   # "Pto" (every second character)
print(word[1::2] )   # "Pto" (every second character)
print(word[::-1])
```

```python
count = 0
while count < 5:
    print(count)
    count += 1    # never forget this or you get infinite loop
```

```python
for i in range(10):
    if i == 5:
        break       # exit loop entirely
    if i % 2 == 0:
        continue    # skip this iteration, go to next
    print(i)        # prints 1, 3
```

```python
for i, val in enumerate(["a", "b", "c"]):
    print(i, val)   # 0 a, 1 b, 2 c
```

```python
s = "hello"
for i in range(len(s)):
    print(s[i])
```

```python
left, right = 0, len(s) - 1
while left < right:
    # process s[left] and s[right]
    left += 1
    right -= 1
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)
```

```python
def clean(text):
    return text.strip().lower()   # returns new string, caller decides what to do with it

name = "  Alice  "
name = clean(name)   # reassign to capture the result
print(name)          # "alice"
```
