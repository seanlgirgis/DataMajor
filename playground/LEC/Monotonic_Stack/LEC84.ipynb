### Approach hint (English only, no code)

The most efficient solution uses a **monotonic increasing stack** to keep track of bar indices.  
As you scan left to right:  
- While the current bar is shorter than the bar at the top of the stack, pop the taller bars and calculate the area they can form (using the current bar as the right boundary, and the previous bar in stack as the left boundary).  
- The width is determined by the difference in indices.  
- After processing all bars, repeat the pop process for any remaining bars in the stack (treating the end of array as height 0).  

This runs in O(n) time because each bar is pushed and popped at most once.

Good luck implementing it yourself!

````python
from typing import List

def largest_rectangle(heights: List[int]) -> int:
    # Edge case â€” empty list
    if not heights:
        return 0

    max_area = 0
    stack = []  # stores (height, index) pairs â€” always in increasing height order
    n = len(heights)

    for i, height in enumerate(heights):

        # Current bar is shorter than the top of stack â€” time to settle accounts
        # Pop everything taller than current height and calculate their max rectangle
        while stack and height < stack[-1][0]:
            h, idx = stack.pop()

            # Left wall  = new stack top after popping (first bar shorter than h to the left)
            # Right wall = current i (first bar shorter than h to the right)
            # If stack is empty, rectangle stretches all the way to the beginning
            area = h * (i - stack[-1][1] - 1) if stack else h * i
            max_area = max(max_area, area)

        stack.append((height, i))

    # Cleanup â€” bars still in stack never found a shorter bar to their right
    # So their right wall is the end of the histogram (n)
    while stack:
        h, idx = stack.pop()
        area = h * (n - stack[-1][1] - 1) if stack else h * n
        max_area = max(max_area, area)

    return max_area


# === TEST CASES ===

# Standard cases
assert largest_rectangle([2, 1, 5, 6, 2, 3]) == 10   # classic example
assert largest_rectangle([2, 4])             == 4    # two bars
assert largest_rectangle([2, 1])             == 2    # answer is single bar
assert largest_rectangle([1, 8, 6, 2, 5, 4, 8, 3, 7]) == 16

# Edge cases
assert largest_rectangle([])                 == 0    # empty list
assert largest_rectangle([5])               == 5    # single bar
assert largest_rectangle([1, 1, 1, 1])      == 4    # all same height
assert largest_rectangle([5, 4, 3, 2, 1])   == 9    # strictly decreasing
assert largest_rectangle([1, 2, 3, 4, 5])   == 9    # strictly increasing
assert largest_rectangle([0, 0, 0])         == 0    # all zeros

print("All tests passed! âœ…")
````
