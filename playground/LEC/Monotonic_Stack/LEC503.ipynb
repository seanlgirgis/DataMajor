## Cell 1: Problem Statement

### Problem Statement
Given a **circular** integer array `nums` (the next element of `nums[i]` is `nums[(i + 1) % nums.length]`), return *the **next greater number** for every element in the array*.

The **next greater number** of a number `x` is the first greater number to its **traversing-order next** in the array, which means you could search circularly to find its next greater number.  
If it doesn't exist, return `-1` for this number.

**Constraints**
- 1 â‰¤ nums.length â‰¤ 10â´
- -10â¹ â‰¤ nums[i] â‰¤ 10â¹

**Examples**  
**Example 1**  
Input: `nums = [1,2,1]`  
Output: `[2,-1,2]`  
Explanation:  
- First 1 â†’ next greater is 2  
- 2 â†’ no greater element (circular search ends at nothing bigger) â†’ -1  
- Second 1 â†’ next greater is 2 (wrapping around)

**Example 2**  
Input: `nums = [1,2,3,4,3]`  
Output: `[2,3,4,-1,4]`  

**Example 3**  
Input: `nums = [5,4,3,2,1]`  
Output: `[-1,-1,-1,-1,-1]`  

**Example 4**  
Input: `nums = [1]`  
Output: `[-1]`

**Approach hint (English only, no code)**  
Use a **monotonic decreasing stack** like in 496, but because it's circular:  
- Traverse the array **twice** (simulate circular behavior)  
- Or append a copy of the array to itself and only consider the first n answers  
- Keep track of **indices** in the stack (not values), so you can fill the result array later  
- Every element is pushed and popped at most twice â†’ still O(n)

Good luck â€” this is a very common follow-up to 496!

````python
from typing import List

def nextGreaterElements(nums: List[int]) -> List[int]:
    """
    Monotonic stack â€“ circular version (O(n) time, O(n) space)
    
    Two common clean ways:
    1. Traverse 2Ã—n times (i % n), using indices
    2. Append nums to itself â†’ traverse once, but only fill first n answers
    
    Here we use the index + modulo style (cleaner in many opinions)
    """
    if not nums:
        return []
    
    n = len(nums)
    result = [-1] * n
    stack = []  # stores indices (decreasing values)
    
    # We simulate two passes by going 0 â†’ 2n-1
    for i in range(2 * n):
        curr_idx = i % n
        curr_val = nums[curr_idx]
        
        # Pop all elements smaller than current value
        while stack and nums[stack[-1]] < curr_val:
            prev_idx = stack.pop()
            result[prev_idx] = curr_val
        
        # Only push index in the first n (to avoid duplicate pushes)
        if i < n:
            stack.append(curr_idx)
    
    return result


````

````python
"""

Alternative style (array doubling â€“ also very common):

"""
    # Alternative (array doubling)
    n = len(nums)
    result = [-1] * n
    stack = []
    
    extended = nums + nums  # simulate circular
    
    for i in range(len(extended)):
        while stack and extended[stack[-1]] < extended[i]:
            idx = stack.pop()
            if idx < n:           # only care about first n
                result[idx] = extended[i]
        
        if i < n:
            stack.append(i)
    
    return result
````
