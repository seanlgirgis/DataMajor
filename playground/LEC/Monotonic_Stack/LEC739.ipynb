### Problem Statement

Given an array of integers `temperatures` representing the daily temperatures, return an array `answer` such that:

`answer[i]` = number of days you have to wait after the ith day to get a **warmer** temperature.  
If there is no future day with a warmer temperature, set `answer[i] = 0`.

**Constraints**  
- 1 ≤ temperatures.length ≤ 30,000  
- 30 ≤ temperatures[i] ≤ 100

**Examples**

**Example 1**  
Input: `temperatures = [73,74,75,71,69,72,76,73]`  
Output: `[1,1,4,2,1,1,0,0]`

**Example 2**  
Input: `temperatures = [30,40,50,60]`  
Output: `[1,1,1,0]`

**Example 3**  
Input: `temperatures = [30,60,90]`  
Output: `[1,1,0]`

### Approach hint (English only, no code)

The most efficient way people usually solve this problem uses a **stack** that keeps track of days (indices) in decreasing order of temperature. As you scan the days from left to right, whenever you find a temperature warmer than some previous days still waiting on the stack, you can calculate how many days those previous days had to wait — and remove them from waiting.

This achieves linear time because each index is pushed and popped from the stack at most once.

Good luck implementing it yourself!

````python
# Write code in plain English first
temps1 = [73, 74, 75, 71, 69, 72, 76, 73]
expected1 = [1, 1, 4, 2, 1, 1, 0, 0]
print ("Input Tempratures     " , temps1   )
print ("expected output       " , expected1)
#start Function 
print ("Function code start here daily_temperatures")
# my param
temps = temps1
result = [0] * len (temps)
# stack to store ids of days where tempratures yet to find warmer days
stack = []
# Loop on temps.. Keep index and cur

for i , cur in enumerate(temps):
    # Only if our stack has some older days to compare to 
    # see we compare today to the past .. we go in the past and if today is hotter than a past day.. calculate the number of days
    # this past day has to wait to meet today which is hotter
    while stack and cur > temps[stack[-1]]:         #Today is hotter we can process the past
        thatDay = stack.pop()
        result[thatDay] = i - thatDay               # Calculating the distance from thatDay to today the hotter day
    #regardless we have to store the id of the current day
    #we do not know what future will be
    stack.append(i)
print ("actual                " , result)
````

````python
from typing import List

# Paste your own implementation here ↑↑↑
def daily_temperatures(temps: List[int]) -> List[int]:
    result = [0] * len (temps)
    stack = []
    for i , cur in enumerate(temps):
        while stack and cur > temps[stack[-1]]:         #Today is hotter we can process the past
            thatDay = stack.pop()
            result[thatDay] = i - thatDay               # Calculating the distance from thatDay to today the hotter day
        stack.append(i)
    return result


        
    


    
````
