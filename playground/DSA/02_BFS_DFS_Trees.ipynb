# Tree Traversal: BFS vs DFS

Welcome! In this notebook, we'll break down the concepts of **Breadth-First Search (BFS)** and **Depth-First Search (DFS)** for trees.

Imagine a tree structure (like a family tree or an organization chart). If you want to visit every node in that tree to find a specific piece of data, there are two main strategies:
1. **BFS**: Go level by level (horizontal).
2. **DFS**: Go as deep as possible down one branch before backing up (vertical).

````python
# First, let's define a simple Tree Node class
from __future__ import annotations       #Python 3.10
class TreeNode:
    def __init__(self, val=0, left:TreeNode=None, right:TreeNode=None):
        self.val = val
        self.left:TreeNode  = left
        self.right:TreeNode = right

# Let's build a simple tree:
#        1
#       / \
#      2   3
#     / \   \
#    4   5   6

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)

print("Tree built successfully!")
````

---
## 1. Breadth-First Search (BFS)

**The Concept:** Explore the tree **level by level** from top to bottom, left to right. 
- Look at level 1 (the root).
- Then look at all nodes on level 2.
- Then all nodes on level 3, etc.

**How it works (The Data Structure):** BFS uses a **Queue** (First-In, First-Out). Think of a line at a grocery store. 
1. Put the root node in the queue.
2. While the queue has people in it: take the first person out, process them, and put their children at the back of the line.

````python
from collections import deque

def bfs_traversal(root):
    if not root:
        return []
    
    result = []
    queue = deque([root]) # 1. Start the queue with the root
    
    while queue:
        # 2. Pop from the FRONT of the line (left side)
        current_node = queue.popleft()
        result.append(current_node.val)
        
        # 3. Add children to the BACK of the line (right side)
        if current_node.left:
            queue.append(current_node.left)
        if current_node.right:
            queue.append(current_node.right)
            
    return result

print("BFS Order:", bfs_traversal(root))
# Expected: [1, 2, 3, 4, 5, 6]
````

## 2. Depth-First Search (DFS)

**The Concept:** Pick a direction (usually left) and keep going down that path until you hit a dead end (a leaf node). Once you hit a dead end, backtrack up one step and try the other path.

**How it works (The Data Structure):** DFS uses a **Stack** (Last-In, First-Out). Think of a stack of plates. 
DFS can be written iteratively (using a list as a stack) or recursively (using the computer's built-in call stack). We will show the recursive way, as it is the most common and elegant for trees.

There are 3 main types of DFS for binary trees depending on *when* you process the node:
1. **Pre-order** (Process Node, then Left, then Right)
2. **In-order** (Left, then Process Node, then Right)
3. **Post-order** (Left, then Right, then Process Node)

````python
# Let's implement DFS Pre-order recursively

def dfs_preorder(node, result):
    if not node:
        return
    
    # 1. Process the current node FIRST (Pre)
    result.append(node.val)
    
    # 2. Drill all the way down the LEFT side
    dfs_preorder(node.left, result)
    
    # 3. Then drill down the RIGHT side
    dfs_preorder(node.right, result)

dfs_result = []
dfs_preorder(root, dfs_result)
print("DFS Pre-order:", dfs_result)
# Order: 1 -> goes left to 2 -> goes left to 4 -> backtracks to 2 -> goes right to 5 ...
# Expected: [1, 2, 4, 5, 3, 6]
````

Great catch â€” they are **NOT** the same order! Look closely:

```

DFS: [1, 2, 4, 5, 3, 6]   â† drill deep first (deep)
```

## Summary cheat sheet: BFS vs DFS

| Feature | BFS (Breadth-First) | DFS (Depth-First) |
| :--- | :--- | :--- |
| **Analogy** | Searching a building floor by floor. | Searching a maze by keeping your hand on the left wall until a dead end. |
| **Data Structure** | Queue (FIFO) | Stack (LIFO) or Recursion |
| **Time Complexity** | **O(N)** - must visit every node | **O(N)** - must visit every node |
| **Space Complexity**| **O(W)** - where W is the widest level of the tree. | **O(H)** - where H is the height of the tree. |
| **When to use?** | Finding the **shortest path** in unweighted graphs. Level-order data. | Exploring all possibilities (like a maze), finding paths, topological sorting. |
