# Graphs: The Basics and Representations

Welcome! In this notebook, we'll cover the core concepts of **Graphs** and how we represent them in code.

A graph is a data structure consisting of **Nodes** (also called Vertices) and **Edges** (the connections between them). They are used to model networks like social networks, city maps, and the internet.

---
## 1. Graph Terminology

- **Vertex (Node)**: A point in the graph.
- **Edge**: A connection between two vertices.
- **Directed vs Undirected**: 
  - In an undirected graph, edges are two-way (like a road). A - B means you can go from A to B and B to A.
  - In a directed graph (Digraph), edges have a direction (like a one-way street). A -> B means you can only go from A to B.
- **Weighted vs Unweighted**: Edges can have "weights" or costs (e.g., miles between cities).

How does a computer understand a graph? There are two main ways: **Adjacency Matrix** and **Adjacency List**.

---
## 2. Representation 1: Adjacency Matrix

An adjacency matrix is a 2D array (a grid) of sizes V x V (where V is the number of vertices).
- Cost: `O(V^2)` space.
- Looking up if an edge exists between `i` and `j` is fast: `O(1)`.
- Iterating over all neighbors of a node is slow: `O(V)`.

````python
# Example of an Undirected Graph's Adjacency Matrix
# Graph:
#  0 -- 1
#  |  /
#  2

# Nodes: 0, 1, 2
matrix = [
    [0, 1, 1],  # Node 0 is connected to 1 and 2
    [1, 0, 0],  # Node 1 is connected to 0
    [1, 0, 0]   # Node 2 is connected to 0 (Notice it's symmetrical for undirected graphs!)
]

def has_edge(u, v):
    return matrix[u][v] == 1

print("Edge between 0 and 1?", has_edge(0, 1))
print("Edge between 1 and 2?", has_edge(1, 2))
````

---
## 3. Representation 2: Adjacency List

An adjacency list represents a graph as an array of linked lists (or a Dictionary of Lists in Python).
- Cost: `O(V + E)` space. (Much more space-efficient for most "sparse" real-world graphs!).
- Looking up a specific edge `(i, j)` takes `O(degree(i))`.
- It's incredibly fast to iterate over a node's neighbors.

````python
# Example of the same Undirected Graph using an Adjacency List
# Node -> [List of neighbors]

graph = {
    0: [1, 2],
    1: [0],
    2: [0]
}

# To find neighbors, we simply access the dictionary key:
print("Neighbors of Node 0:", graph[0])
print("Neighbors of Node 1:", graph[1])
````

---
## Summary
**Always use an Adjacency List** unless you have a specific reason not to (like an extremely dense graph where almost every node is connected to every other node).
