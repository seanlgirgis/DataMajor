```python
words = ["banana", "apple", "cherry","Orange","date"]
words.sort(key=len)      # sort by length
print(words)
```

```python
pairs = [(1,'b'), (2,'a')]
pairs.sort(key=lambda x: x[0])  # sort by second element
print(pairs)
pairs.sort(key=lambda x: x[1])  # sort by second element
print(pairs)
```
