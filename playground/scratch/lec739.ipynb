```python
def daily_temperatures(temps):
    stack = []      # stores indices
    result = [0] * len(temps)
    
    for i, temp in enumerate(temps):
        while stack and temps[stack[-1]] < temp:
            idx = stack.pop()
            result[idx] = i - idx
        stack.append(i)
    
    return result
```
